```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Dataset
items = [
    "action movie hero fight",
    "romantic love story",
    "action adventure hero",
    "comedy fun laughter",
    "romantic emotional drama"
]

# Convert text to vectors
cv = CountVectorizer()
matrix = cv.fit_transform(items)

# Compute similarity
similarity = cosine_similarity(matrix)

def recommend(index):
    scores = list(enumerate(similarity[index]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print("\nSelected:", items[index])
    print("Recommended:")
    
    for i in scores[1:3]:
        print("-", items[i[0]])

# Run example
recommend(0)
```
