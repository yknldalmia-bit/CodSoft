```python
import random

board = [" "]*9

def print_board():
    for i in range(0, 9, 3):
        print(board[i], "|", board[i+1], "|", board[i+2])
    print()

def check_win(player):
    wins = [(0,1,2),(3,4,5),(6,7,8),
            (0,3,6),(1,4,7),(2,5,8),
            (0,4,8),(2,4,6)]
    
    return any(board[a]==board[b]==board[c]==player for a,b,c in wins)

def player_move():
    while True:
        move = int(input("Enter position (0-8): "))
        if board[move] == " ":
            board[move] = "X"
            break
        else:
            print("Invalid move!")

def ai_move():
    empty = [i for i in range(9) if board[i]==" "]
    move = random.choice(empty)
    board[move] = "O"
    print("AI chose:", move)

def game():
    for turn in range(9):
        print_board()
        
        if turn % 2 == 0:
            player_move()
            if check_win("X"):
                print_board()
                print("🎉 You Win!")
                return
        else:
            ai_move()
            if check_win("O"):
                print_board()
                print("🤖 AI Wins!")
                return

    print_board()
    print("Draw!")

game()
```
